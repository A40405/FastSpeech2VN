# FastSpeech2VN on Kaggle (MFA + IPA)This notebook uses the new IPA-based Vietnamese pipeline for FastSpeech2VN: download data, prepare raw files, build MFA assets, train/export TextGrid, optionally upload/download TextGrid from Hugging Face, preprocess, train, and infer WAV.Recommended accelerator: **2 x T4**.Important note: this notebook uses the repo's IPA-style Vietnamese frontend and an MFA workflow inspired by ViMFA. It does not bundle a pretrained ViMFA G2P model or pretrained ViMFA acoustic model.This notebook creates a separate micromamba environment for MFA under `/kaggle/working`, so the MFA setup can take a while on Kaggle.

```python
# 1) Clone repo!git clone https://github.com/A40405/FastSpeech2VN.git%cd /kaggle/working/FastSpeech2VN
```

```python
# 2) Check GPU!nvidia-smi
```

```python
# 3) Install Python packages for the repo!python -m pip install --upgrade pip!pip install -r requirements-kaggle.txt
```

```python
# 4) Install micromamba and create a dedicated MFA env!/bin/bash -lc 'set -euo pipefail; cd /kaggle/working; mkdir -p mm; curl -Ls https://micro.mamba.pm/api/micromamba/linux-64/latest | tar -xvj -C mm bin/micromamba --strip-components=1; /kaggle/working/mm/micromamba create -y -p /kaggle/working/mfa-env -c conda-forge montreal-forced-aligner'
```

```python
# 5) Check MFA!/bin/bash -lc '/kaggle/working/mm/micromamba run -p /kaggle/working/mfa-env mfa version'
```

```python
# 6) Download InfoRe1 dataset!/bin/bash -lc 'rm -rf /kaggle/working/FastSpeech2VN/corpus/infore1_25hours /kaggle/temp/infore1_downloads; python scripts/download_infore1_dataset.py --dataset doof-ferb/infore1_25hours --split train --output-dir ./corpus/infore1_25hours --download-dir /kaggle/temp/infore1_downloads --cleanup-shards'
```

```python
# 7) Prepare raw files (.wav, .lab, .phones)!python prepare_align.py ./config/InfoRe1_25hours/preprocess.yaml
```

```python
# 8) Build MFA corpus and lexicon from the repo phonemizer!python scripts/build_infore1_mfa_assets.py --raw-root ./raw_data/InfoRe1 --corpus-root ./mfa_corpus/InfoRe1 --lexicon-path ./mfa_assets/infore1_vi.dict
```

```python
# 9) Train MFA aligner and export TextGrid!python scripts/run_mfa_train_alignment.py --mfa /kaggle/working/mfa-env/bin/mfa --corpus-root ./mfa_corpus/InfoRe1 --lexicon-path ./mfa_assets/infore1_vi.dict --output-root ./preprocessed_data/InfoRe1/TextGrid --model-path ./mfa_assets/infore1_vi_acoustic_model.zip --num-jobs 4 --single-speaker --overwrite
```

```python
# 10) Upload TextGrid to Hugging Faceimport osfrom huggingface_hub import HfApirepo_id = 'YOUR_USERNAME/infore1-textgrid'local_folder = './preprocessed_data/InfoRe1/TextGrid'token = os.environ.get("HF_TOKEN")assert token, "Set HF_TOKEN in the Kaggle environment first."api = HfApi(token=token)api.create_repo(repo_id=repo_id, repo_type="dataset", private=True, exist_ok=True)api.upload_folder(    folder_path=local_folder,    repo_id=repo_id,    repo_type="dataset",    path_in_repo="TextGrid",)print(f"Uploaded {local_folder} to https://huggingface.co/datasets/{repo_id}")
```

```python
# 11) Download TextGrid from Hugging Faceimport osfrom huggingface_hub import snapshot_downloadrepo_id = 'YOUR_USERNAME/infore1-textgrid'target_dir = './preprocessed_data/InfoRe1/TextGrid'token = os.environ.get("HF_TOKEN")assert token, "Set HF_TOKEN in the Kaggle environment first."snapshot_download(    repo_id=repo_id,    repo_type="dataset",    local_dir=target_dir,    local_dir_use_symlinks=False,    token=token,)print(f"Downloaded to {target_dir}")
```

```python
# 12) Preprocess acoustic features!python preprocess.py ./config/InfoRe1_25hours/preprocess.yaml
```

```python
# 13) Train FastSpeech2!python train.py -p ./config/InfoRe1_25hours/preprocess.yaml -m ./config/InfoRe1_25hours/model.yaml -t ./config/InfoRe1_25hours/train.yaml
```

```python
# 14) Download HiFi-GAN pretrained weights!python scripts/download_hifigan_pretrained.py
```
